```python
import pandas as pd
import anthropic
```

```python
def api(userInput):
    client = anthropic.Anthropic(api_key="")
    message = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=1000,
        messages=[
            {
                "role": "user",
                "content": "Give me a list of most important/essential keywords (at most 20) from this job description(each keyword seperated by comma):"+userInput
            }
        ]
    )
    return (message.content)
```

```python
df = pd.read_csv('postings.csv')
print(df.columns)
df_sample = df.sample(n=20, random_state=42)
```

```python
df_sample_pruned = df_sample[['title', 'description', 'location', 'application_url']]
df_sample_pruned = df_sample_pruned.reset_index(drop=True)
df_sample_pruned
```

```python
now = [] 
for i in range(20):
    now.append(api(df_sample_pruned['title'][i] + 'DESCRIPTION: ' + df_sample_pruned['description'][i]))
    print(i)

df_sample_pruned['keywords'] = now 
df_sample_pruned
```

```python
for i in range(20):
    # df_sample_pruned['keywords'][i] = df_sample_pruned['keywords'][i][0].text
    df_sample_pruned['keywords'][i] = [k.strip() for k in df_sample_pruned['keywords'][i].split(',')]
df_sample_pruned
```

```python
df_sample_pruned.to_json('df_sample_pruned.json', orient='records', lines=False)
```
