```python
def api(userInput):
    client = anthropic.Anthropic(
        api_key="")
    message = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=1000,
        messages=[
            {
                "role": "user",
                "content": "Give me a list of most important/essential keywords (at most 20) from this job description(each keyword seperated by comma):"+userInput
            }
        ]
    )
    return (message.content)
```
